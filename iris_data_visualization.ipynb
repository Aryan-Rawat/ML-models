```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('/content/iris-write-from-docker.csv')
dataset.head(10)
```

```python
dataset.info()
```

```python
dataset['class'].value_counts()
```

```python
dataset.describe()
```

```python
plt.bar(dataset['class'].unique(), dataset['class'].value_counts(),
        width = 0.8, color = ['black', 'green'])
plt.title('Bar Chart of class value', fontsize = 15)
plt.xlabel('Class Names', fontsize = 12)
plt.ylabel('Number of sets in class', fontsize = 12)

plt.show()
```

```python
x = dataset.sepal_length
y = dataset.petal_length
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.plot(x, y)
```

```python
x = dataset.sepal_length
y = dataset.petal_length
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.scatter(x, y)

plt.figure(figsize = (10, 10))
plt.subplot(2,2,1)
plt.scatter(dataset.petal_width, dataset.petal_length, color = 'y', marker = 'v', label = 'Petal Width vs Petal Length')
plt.legend()
plt.show()
```

```python
plt.pie(dataset['class'].value_counts(), labels = dataset['class'].unique(), radius = 2, autopct = '%1.1f%%', explode = [0.2, 0.2, 0.2])
plt.title('Pie Chart of class values', fontsize = 15)
plt.show()
```
